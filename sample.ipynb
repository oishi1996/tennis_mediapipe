```python
import cv2
import mediapipe as mp
import cv2
import mediapipe as mp
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
mp_pose = mp.solutions.pose
mp_drawing = mp.solutions.drawing_utils
```

```python
video_path = 'data/so_serve_1.mp4'  # 動画のパスを指定
```

```python


# MediaPipe Poseモジュールのセットアップ
mp_pose = mp.solutions.pose
mp_drawing = mp.solutions.drawing_utils

# 動画の読み込み
#video_path = 'tennis_serve.mp4'  # 動画のパスを指定
cap = cv2.VideoCapture(video_path)

# 3Dグラフの初期化
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Poseモジュールの初期化
with mp_pose.Pose(min_detection_confidence=0.5, min_tracking_confidence=0.5) as pose:
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            print("動画の読み込みが終了しました")
            break

        # BGR画像をRGBに変換
        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        image.flags.writeable = False

        # 骨格推定
        results = pose.process(image)

        # 画像に骨格ランドマークを描画
        image.flags.writeable = True
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)

        if results.pose_world_landmarks:
            # 3Dランドマークを取得
            landmarks = results.pose_world_landmarks.landmark
            
            # 各ランドマークのx, y, z座標を抽出
            xs = [landmark.x for landmark in landmarks]
            ys = [landmark.y for landmark in landmarks]
            zs = [landmark.z for landmark in landmarks]

            # 3D空間に描画
            ax.clear()
            ax.scatter(xs, ys, zs, c='r', marker='o')

            # 軸ラベルを設定
            ax.set_xlabel('X')
            ax.set_ylabel('Y')
            ax.set_zlabel('Z')

            # ランドマークの番号を付与
            for i, (x, y, z) in enumerate(zip(xs, ys, zs)):
                ax.text(x, y, z, f'{i}', color='blue')

            # グラフの表示を更新
            plt.draw()
            plt.pause(0.001)

        # 結果を表示
        cv2.imshow('Tennis Serve Pose Estimation', image)

        if cv2.waitKey(10) & 0xFF == ord('q'):
            break

cap.release()
cv2.destroyAllWindows()
```

```python
# 動画の読み込み
cap = cv2.VideoCapture(video_path)

# Poseモジュールの初期化
with mp_pose.Pose(min_detection_confidence=0.5, min_tracking_confidence=0.5) as pose:
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            print("動画の読み込みが終了しました")
            break

        # BGR画像をRGBに変換
        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        image.flags.writeable = False

        # 骨格推定
        results = pose.process(image)

        # 画像に骨格ランドマークを描画
        image.flags.writeable = True
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)

        if results.pose_landmarks:
            # 2Dランドマークを描画
            mp_drawing.draw_landmarks(
                image, results.pose_landmarks, mp_pose.POSE_CONNECTIONS)

            # 3Dランドマークを取得して出力
            if results.pose_world_landmarks:
                print("3Dランドマーク情報:")
                for idx, landmark in enumerate(results.pose_world_landmarks.landmark):
                    print(f"ランドマーク {idx}: x={landmark.x}, y={landmark.y}, z={landmark.z}")

        # 結果を表示
        cv2.imshow('Tennis Serve Pose Estimation', image)

        if cv2.waitKey(10) & 0xFF == ord('q'):
            break

cap.release()
cv2.destroyAllWindows()
```
